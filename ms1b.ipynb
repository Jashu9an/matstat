```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
N = 100
tt = 135
np.random.seed(44)
X = np.random.randint(0, tt, N)
```

```python
X
```

```python
tt1 = [2*np.sum(X[:n])/n for n in range(1,N+1)]
tt2 = [(n+1)*np.min(X[:n]) for n in range(1,N+1)]
tt3 = [(n+1)*np.max(X[:n])/n for n in range(1,N+1)]
tt4 = [(np.min(X[:n]) + np.max(X[:n])) for n in range(1,N+1)]
tt5 = [(X[0] + 1/(n-1)*np.sum(X[1:n])) for n in range(2,N+1)]
tt5.insert(0, X[0])
```

```python
fig, ax = plt.subplots(figsize = (30, 15))
x = np.arange(1, N+1, dtype = int)
y1 = [abs(s - tt) for s in tt1]
y2 = [abs(s - tt) for s in tt2]
y3 = [abs(s - tt) for s in tt3]
y4 = [abs(s - tt) for s in tt4]
y5 = [abs(s - tt) for s in tt5]

plt.grid()
ax.plot(x, y1, 'black')
ax.plot(x, y2, 'red')
ax.plot(x, y3, 'blue')
ax.plot(x, y4, 'purple')
ax.plot(x, y5, 'yellow')

plt.show()
```
